## 3. Initial Observations

- The dataset contains information related to customers, vehicles, policies, claims, and payments.
- The dataset spans four years (2015–2018).
- There is no explicit **Renewed** or **Churn** column.
- The **Lapse** column represents the number of policies previously cancelled or lapsed due to non-payment and appears to be a historical feature rather than the prediction target.
- Several date columns may help reconstruct customer renewal history.
- Further investigation is required to determine whether a valid target variable can be engineered from the available data.
