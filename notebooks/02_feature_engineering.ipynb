Importing libraries and loading dataset

```python
from pathlib import Path
import pandas as pd
import numpy as np

DATA_PATH = Path("../data/raw/Motor vehicle insurance data.csv")

df = pd.read_csv(DATA_PATH, sep = ';')

df_fe = df.copy()
```

```python
date_coloumns = ["Date_start_contract","Date_last_renewal","Date_next_renewal","Date_birth","Date_driving_licence","Date_lapse","Year_matriculation"]
```

```python
#Converting Dates
for col in date_coloumns:
    df_fe[col] = pd.to_datetime(df_fe[col],errors='coerce')
```

```python
#Customers Age
Reference_year = 2018
df_fe["Customer_Age"] = (Reference_year - df_fe['Date_birth'].dt.year)
```

```python
#Driving Experience
df_fe["Driving_Experience"] = (Reference_year - df_fe['Date_driving_licence'].dt.year)
```

```python
#Customer Tenure
df_fe["Customer_Tenure"] = (Reference_year - df_fe['Date_start_contract'].dt.year)
```

```python
#Vehicle Age
df_fe["Vehicle_Age"] = (Reference_year - df_fe["Year_matriculation"].dt.year)
```

```python
#claim frequency
df_fe["Claim_per_year"] = df_fe["N_claims_year"] / df_fe["Seniority"]
```
